```python
def download_book(book, url):
    #book = row["Book Title"]
    #url = row["OpenURL"]
    fp = urllib.request.urlopen(url)
    mybytes = fp.read()

    html = mybytes.decode("utf8")
    fp.close()
    
    start = html.find("/content/pdf/")
    substr = html[start:]
    end = substr.find(".pdf")
    
    path = "https://link.springer.com/" + substr[:end] +".pdf"
    print(book)
    print(path)
    r = requests.get(path, stream=True)
    
    with open("{}.pdf".format(book.replace(" ", "_")), "wb") as fd:
        for chunk in r.iter_content(2000):
            fd.write(chunk)
    
    #wget.download(url, "books\\"+book.replace(" ", "_")+".pdf")
    
```

```python
download_book('Fundamentals of Power Electronics', 'http://link.springer.com/openurl?genre=book&isbn=978-0-306-48048-5')
```

```python
pandita.apply(download_book, axis=1)
```
